```python
def min_max_scale(X):
    X_min = X.min(axis=0)
    X_max = X.max(axis=0)
    return (X - X_min) / (X_max - X_min)

X_scaled = min_max_scale(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
# Color maps for plotting
cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
cmap_light = ListedColormap(["#FFBBBB", "#BBFFBB", "#BBBBFF"])
```

```python
# --- Activation Functions ---
def linear(H):
    return H

def ReLU(H):
    return H * (H > 0)

def sigmoid(H):
    return 1 / (1 + np.exp(-H))

def softmax(H):
    eH = np.exp(H - np.max(H, axis=1, keepdims=True))  # Stability fix
    return eH / eH.sum(axis=1, keepdims=True)
```

```python
# --- Loss Functions ---
def cross_entropy(Y, P_hat):
    return -(1 / len(Y)) * np.sum(Y * np.log(P_hat + 1e-10))  # stability

def OLS(Y, Y_hat):
    return (1 / (2 * len(Y))) * np.sum((Y - Y_hat) ** 2)

# --- Useful functions ---
def one_hot(y):
    N = len(y)
    K = len(set(y))
    Y = np.zeros((N, K))
    Y[np.arange(N), y] = 1
    return Y

def accuracy(y, y_hat):
    return np.mean(y == y_hat)

def R2(y, y_hat):
    return 1 - np.sum((y - y_hat) ** 2) / np.sum((y - y.mean()) ** 2)

def derivative(Z, a):
    if a == linear:
        return 1
    elif a == sigmoid:
        return Z * (1 - Z)
    elif a == np.tanh:
        return 1 - Z ** 2
    elif a == ReLU:
        return (Z > 0).astype(int)
    else:
        raise ValueError("Unknown Activation")
```

```python
# --- ANN Class ---
class ANN:
    def __init__(self, architecture, activations=None, mode=0):
        self.mode = mode
        self.architecture = architecture
        self.activations = activations
        self.L = len(architecture) + 1  # Total layers incl. output

    def fit(self, X, y, eta=1e-3, epochs=1e3, show_curve=False):
        N, D = X.shape
        epochs = int(epochs)
        Y = y if self.mode else one_hot(y)
        K = 1 if self.mode else Y.shape[1]

        # Initialize weights and biases
        self.W = {
            l: np.random.randn(input_size, output_size) * np.sqrt(2 / input_size)
            for l, (input_size, output_size) in enumerate(zip([D] + self.architecture, self.architecture + [K]), 1)
        }
        self.B = {l: np.zeros(output_size) for l, output_size in enumerate(self.architecture + [K], 1)}

        if self.activations is None:
          self.a= {l: ReLU for l in range(1, self.L)}
        else:
          self.a={l: act for l,act in enumerate(self.activations, 1)}
    
        #Output Activation Functions
        if self.mode:
          self.a[self.L]=linear
        else:
          self.a[self.L]=softmax


        J = np.zeros(epochs)

        #SGD Progression
        for epoch in range(epochs):
            self.__forward__(X)

            if self.mode:
                J[epoch] = OLS(Y, self.Z[self.L])
            else:
                J[epoch] = cross_entropy(Y, self.Z[self.L])

            dH = (1 / N) * (self.Z[self.L] - Y)

            for l in sorted(self.W.keys(), reverse=True):

                dW = self.Z[l-1].T@dH
                dB = dH.sum(axis=0)
        
                self.W[l] -= eta*dW
                self.B[l] -= eta*dB

                if l>1:
                  dZ =dH@self.W[l].T
                  dH = dZ*derivative(self.Z[l-1], self.a[l-1])

        if show_curve:
            plt.figure()
            plt.plot(J)
            plt.xlabel("Epochs")
            plt.ylabel(r"$\mathcal{J}$")
            plt.title("Training Curve")
            plt.grid(True)
            plt.show()

    def __forward__(self, X):
        self.Z = {0: X}
        for l in range(1, self.L + 1):
            self.Z[l] = self.a[l](self.Z[l - 1] @ self.W[l] + self.B[l])

    def predict(self, X):
        self.__forward__(X)
        if self.mode:
          return self.Z[self.L]
        else:
          return self.Z[self.L].argmax(axis=1)
    
```

```python
def main_class():
    
    X_train_2d = X_train.values[:, :2]
    y = y_train   
    print("Input shape:", X_train_2d.shape)
    plt.figure()
    plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y, s=6, alpha=0.6)
    plt.xlabel("Age (scaled)")
    plt.ylabel("Fare (scaled)")
    plt.title("Titanic Dataset: Age vs Fare colored by Survival")
    plt.show()
    
    # Train the ANN model ( 2 features)
    my_ann_classifier = ANN(architecture= [4,2], activations= [np.tanh,ReLU])
    my_ann_classifier.fit(X_train_2d, y_train, eta=3e-2,epochs = 1e4, show_curve=True)
    y_hat = my_ann_classifier.predict(X_train_2d)
    print(f"Training Accuracy: {accuracy(y_train, y_hat):0.4f}")
    print(my_ann_classifier.W)
    print(my_ann_classifier.B)

    # Visualize decision boundary 
    x1 = np.linspace(X_train_2d[:, 0].min() - 1, X_train_2d[:, 0].max() + 1, 1000)
    x2 = np.linspace(X_train_2d[:, 1].min() - 1, X_train_2d[:, 1].max() + 1, 1000)
    xx1, xx2 = np.meshgrid(x1, x2)
    Z = my_ann_classifier.predict(np.c_[xx1.ravel(), xx2.ravel()]).reshape(*xx1.shape)
    
    # Plot the decision boundary
    plt.figure()
    plt.pcolormesh(xx1, xx2, Z, cmap=cmap_light)
    plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_train, cmap=cmap_bold, alpha=0.2)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    plt.title('ANN Decision Boundary for Titanic Dataset')
    plt.show()
    
    # Final scatter plot of predictions
    plt.figure()
    plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=y_hat, s=6)

def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

if __name__ == "__main__":
    main_class()
```
